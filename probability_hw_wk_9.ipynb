### Below is the sample variance and standard deviation for all numerical columns in the same output.  We use sample variance and not population variance because we are working with a subset of rentable properties in Belize, and we can not be sure that all properties have been captured.  This makes the sample calculation a more accurate one.  My takeaway from each measure is as follows:
 - For some categories like id and host_id these measures do not provide useful information, as these are not scales of measurement at all.
 - For latitude and longitude (measured in degrees) it does indicate that the are covered by the dataframe is realtively small, generally located within a close range of the same latitude and longitude.
 - Some categories indicate a high degree of variability.  Price (measured in currency) indicates that there are wide swings in possible prices, as we saw earlier with the large outliers.  Availability_365 (measured in days) is also a very high standard deviation.  This is because of some outlier rentals that are rarely available.  Calculated_host_listings (measured in number of properties) also has high variability, likely because airbnbs are often a single property owned and rented out by an individual, but some owners may own whole complexes of buildings. Number of reviews (measured in numbers of reviews) has a large standard deviation, likely caused by some properties with large amounts of reviews that have been renting out much longer.
 - Some categories show relatively low variability.  Minimum nights (measured in days) is a low number, with standard deviation of less than a week.  Reviews per month (measured in numbers of reviews) is also a low standard deviation, with less than a full review difference per month. 
 - Generally, the standard deviation is a more useful number to go by, as very few of these measures make sense as differences squared.

```python
pd.DataFrame([np.var(airbnb_df[:], ddof=1), np.std(airbnb_df[:], ddof=1)], index=['Variance', 'Std. dev'])
```
